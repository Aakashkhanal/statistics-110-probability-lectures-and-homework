### Gambler's Ruin Problem:
Two gamblers, A and B, sequence of rounds where they bet a dollar(each rounds are independent), where p = P(A wins a certain round), q = 1-p. Find the probability that A wins the entire game so that B is ruined?  
Assuming A starts with i dollars and B starts with (N-i) dollars. Baically total wealth is N.  
<br>
**Solution:**  
*Strategy* : Condition on the first step.
 
 Let's say $P_i = P(\text{A wins game}|\text{A starts at \$i})$  
 And the extreme conditions are  
 <br>
 \begin{align}
 P_0 &= 0 \\
 P_N &= 1 
 \end{align}
 so,
 <br>
 <br>
 
 \begin{align}
 P_i = p \times P_{i+1} + q \times P_{i-1}, \qquad 1 \leq i \leq N
 \end{align}
 
 <br>
 <br>
 
 **The above equation is known as difference equation**  
 So, let's solve it.  
 There are a lot of deep connecctions between differential and difference equation and one of the ways to solve is to guess.  
let's guess $P_i = x^i$. So, the equation becomes
<br>
<br>
\begin{align}
x^i &= p \times x^{i+1} + q \times x^{i-1}  \\
px^2 - x + q &= 0  \\
x &= \frac{1 \pm \sqrt{1 - 4pq}}{2p}
\end{align}

Again we know that $$p=1-q$$ 
So,
<br>
<br>
\begin{align}
1-4pq &= 1-4p(1-p) \\
&= 4p^2-4p+1 \\
&= (2p-1)^2 \\
\end{align}

So,
<br>
<br>

$$
x =
\begin{cases}
1 \\[2ex]
\frac{q}{p}
\end{cases}
$$

So, the general solution from the value of x is,

\begin{align}
P_i = A 1^i + B (\frac{q}{p})^i && ,p\neq q
\end{align}

So, using the extreme conditions we get the value of A and B. So, finally  
<br>
<br>
$$
P_i =
\begin{cases}
    \frac{1 - (\frac{q}{p})^i}{1 - (\frac{q}{p})^N} && if P \neq q\\[2ex]
? && if p = q
\end{cases}
$$

So, let's solve the case for p = q.

let's take $x = \frac{q}{p}$
\begin{align}
P_i &= \lim\limits_{x \to 1} \frac{1 - x^i}{1 - x^N} &&, \text{it's in 0/0 form} \\
&= \lim\limits_{x \to 1} \frac{ix^{i-1}}{Nx^{N-1}} &&, \text{using L'Hospital Rule} \\
&= \frac{i}{N}
\end{align}
<br>
<br>

##### Few figures using the above calculations
If at a point A and B start with same amount of money i.e $i=N-i$ and probability of A winning each round is slightly lesser than B i.e $p=0.49$. Using the above equtions what we get is,  
<br>
$N = 20 \Rightarrow P(\text{A winning the game}) = 0.40$
$N = 100 \Rightarrow P(\text{A winning the game}) = 0.12$
$N = 200 \Rightarrow P(\text{A winning the game}) = 0.02$

<br>
<br>

So, if we put these ideas in Los Vegas scenario, we will always have lesser money than the casino as well as the games are slighlty unfair to very unfair to the gamblers. That's why it's called as **Gambler's Ruin**, because if you keep on playing you will eventually loose all the money.
<br>

##### Now let's try to understand if there is any possibility that the game can go on for infinity  

If at one point A starts with i and B starts with N-i, so the probability of A winning in fair case i.e $p=q$ is $\frac{i}{N}$ and the probability of B winning is $\frac{N-i}{N}$. And if we add these two we get 1. So, there is no scope where the game goes to infinite loop. So, that case is not possible. That means the there is zero probability that the game goes forever. Same thing can be proved for unbiased case also where, $p \neq q$
<br>

### Ramdom Variables

**What is a Random Variable?**  
It's a function from sample space $\mathbb{S}$ to $\mathbb{R}$. 
To explain that, let's say we have a sample space $S$ and in that sample space we has an event $s$. So the random variable is a function which will map that small event $s$ to a value in Real scale.  
Think of it as a numerical **Summary** of an aspect of the experiment. 

<br>
<br>

##### Random Variable Examples  
**Bernoulli Distribution:**
A random variable $X$ is said to have Bernoulli Distribution, if $X$ has two possible values $0$ and $1$, and if $P(X = 1) = p$, $P(X=0)=1-p$.  
<br>
<br>
**Binomial Distribution(n,p):**
The distribution of number of successes in n independent **Bern(p)** trials is called **Bin(n,p)**, and it's distribution is given by
<br>
<br>
\begin{align}
P(X=k) = {n \choose k} p^k (1-p)^{n-k} && ,0 \leq k \leq n && \text{and k is an integer}
\end{align}

where $X$ is number of successes.

This is also know as **PMF**(Probability Mass Function)
